# Regressão linear aplicada a altura média de um pai e seu filho (a)

## Importing pandas and json to deal with data

```python
import pandas as pd
import json
import numpy as np
```

## Create a json and exporting it as csv file in data folder

```python
database = json.dumps(
    [
        {"X": 164, "Y": 166},
        {"X": 166, "Y": 166},
        {"X": 169, "Y": 171},
        {"X": 169, "Y": 166},
        {"X": 171, "Y": 171},
        {"X": 173, "Y": 171},
        {"X": 173, "Y": 178},
        {"X": 176, "Y": 173},
        {"X": 178, "Y": 178},
    ]
)
database_json_file = pd.read_json(database)
database_json_file.to_csv("./data/height.csv", encoding="utf-8")
```

## Creating a dataframe and reading height.csv

```python
df = pd.read_csv("./data/height.csv")
fathers_series = df["X"]
sons_series = df["Y"]
```

# Plot a relation between fathers and sons

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(fathers_series, sons_series, color="purple")
plt.xlabel("Altura média dos pais X")
plt.ylabel("Altura média dos filhos Y")
plt.show()
```

## Creating a line function

## $\text{Line equation is given by: } y = \beta x + \alpha$


## Finding $\beta$ $\text{value by: } \beta = \dfrac{n \sum_{\left(x_{i}y_{i}\right)} - \left(\sum_{x_{i}} \right) \sum_{y_{i}}}{n \sum{}{x^2_{i}} - \left( \sum_{}{x_{i}}\right)^{2}}$

## Finding $\alpha$ value by: $\alpha =  \dfrac{\sum_{y_{i}} - b\sum_{x_{i}}}{n}$

```python
b = 0.872
a = 22
y_values = [(b * x + a) for x in fathers_series]
```

## Plotting line function

```python
plt.scatter(fathers_series, sons_series, color="purple")
plt.plot(fathers_series, y_values, color="black")
plt.xlabel("Altura média dos pais X")
plt.ylabel("Altura média dos filhos Y")
plt.show()
```
